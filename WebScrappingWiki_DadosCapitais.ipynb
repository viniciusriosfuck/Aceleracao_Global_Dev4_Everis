```
!pip install wikipedia
import wikipedia as wp

import pandas as pd

wp.set_lang("pt")

def import_wiki_table(page, table_nr, print_bool=True):
    """
        given page name and number of a table
        return the correspondent dataframe
    """
    html = wp.page(page).html().encode("UTF-8")
    try:  
        df = pd.read_html(html, encoding='utf-8')[table_nr]
    # Try 2nd table first as most pages contain contents table first
    except IndexError:
        raise("Erro")
    if print_bool:
        print(df.to_string())
    return df
```

```
page = "Lista_de_capitais_do_Brasil_por_população"
df_populacao = import_wiki_table(page, 0)
df_populacao.drop(columns=["Posições em 2020", "População em 2010", "População em 2000",  "Pos. 2000"], inplace=True)
# df_populacao.rename(columns={"População em 2020 (IBGE)":"Prefeito"},inplace=True)
df_populacao['Capitais'] = df_populacao['Capitais'].str.replace(r'\[[^()]*\]', '')
df_populacao["População em 2020 (IBGE)"] = df_populacao["População em 2020 (IBGE)"].str.replace(".","")
df_populacao
```

```
page = "Eleições_municipais_no_Brasil_em_2020"
df_eleicoes2020 = import_wiki_table(page, 4)
df_eleicoes2020
df_eleicoes2020.rename(columns={"Capital":"Capitais"},inplace=True)

col = "Eleito no"
mask = df_eleicoes2020[col].str.startswith("1")
df_eleicoes2020.loc[mask, col] = "2020-11-15"
df_eleicoes2020.loc[~mask, col] = "2020-11-29"
mask = df_eleicoes2020["Capitais"] == "Macapá"
df_eleicoes2020.loc[mask,col] = "2020-12-20"
df_eleicoes2020["Posse"] = "2021-01-01"
df_eleicoes2020
```

```
page = "Lista_dos_maiores_colégios_eleitorais_do_Brasil"
df_eleitores = import_wiki_table(page, 0, print_bool=False)
df_eleitores.rename(columns={"Município":"Capital", "Estado":"UF"},inplace=True)
df_eleitores["Eleitores (hab.)"] = df_eleitores["Eleitores (hab.)"].str.replace(".","")

df = df_populacao.merge(df_eleicoes2020.drop(columns=["Estado"]),on="Capitais")
df.columns = ["Capital", "UF", "Populacao2020", "Prefeito", 
              "Partido", "VicePrefeito", "DataUltimaEleicao", "DataPosse"]
# há Palmas além do TO
df = df.merge(df_eleitores.drop(columns=["Posição"]),on=["Capital","UF"])
# df.drop(columns=["Estado"],inplace=True)
df_merged = df.rename(columns={"Eleitores (hab.)": "Eleitores"})
df_merged
```

```
%%time
lst_UF = list(df_populacao["Unidade federativa"])
lst_capitais = list(df_populacao["Capitais"])

lst_df = []
for page in lst_capitais:
    table_nr = 0
    if page == "São Luís":
        page = "São Luís (Maranhão)"
    elif page == "Belém":
        page = "Belém (Pará)"
    elif page == "Vitória":
        page = "Vitória (Espírito Santo)"
    elif page == "Belém":
        page = "Belém (Pará)"
    elif page == "Natal":
        page = "Natal (Rio Grande do Norte)"
    elif page in ['Maceió', 'Teresina', 'Cuiabá', 'Porto Velho']:
        table_nr = 1
    elif page in ['Boa Vista', 'Rio Branco']:
        table_nr = 2
    print(page)
    lst_df.append(import_wiki_table(page, table_nr,print_bool=False))
        
```

```
lst_fundacao = []
for df in lst_df:
    mask = df.iloc[:,0] == "Fundação"
    s = df.iloc[:,1].loc[mask]
    # capital = s.name
    # fundacao = s
    capital = s.name[:-2]
    fundacao = s.str.replace("\([^()]*\)", " ").str.replace(r'\[[^()]*\]', '').str.strip().item()
    # if type(capital) == str: 
        # try:
        #     pass
        # except:
        #     pass
    lst_fundacao.append([{"Capital": capital}, {"Fundacao": fundacao}])

df_1 = pd.DataFrame([lst[0] for lst in lst_fundacao])
df_2 = pd.DataFrame([lst[1] for lst in lst_fundacao])
df_fundacao = pd.concat([df_1, df_2], axis='columns')

df = df_merged.merge(df_fundacao, on="Capital")
meses=('janeiro','fevereiro','março','abril','maio','junho',
       'julho','agosto','setembro','outubro','novembro','dezembro')
dct_meses = dict(zip(meses,[str(x+1) for x in range(12)]))

col = "Fundacao"
df[col] = df[col].replace(dct_meses, regex=True)
df[col] = df[col].str.replace(" de ","/")
df[col].loc[df[col] == '1907'] = "2/10/1907"
df[col].loc[df["Capital"]=="Curitiba"] = "29/3/1693"
df["AnoFundacao"] = df[col].str[-4:]
for col in ["DataUltimaEleicao", "DataPosse"]:
    df[col] = pd.to_datetime(df[col]).dt.strftime("%d/%m/%Y")
df
```
